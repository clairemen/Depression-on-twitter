```python
import numpy as np
import pandas as pd
from scipy import stats
#正样本
posfilename='G:\\sufe\\mental illness\\Dataset\\labeled\\positive\\data\\feature_count\\pos_count.csv'
pos_df = pd.read_csv(posfilename)

cols = ['a_count','c_count','me_count','you_count','him_count','CC','CD','IN','JJ','MD','NN','RB','RP','VB','WDT','WRB']
for item in cols:
    pos_df[item] /= pos_df['w_count']
    
pos_df['w_count']=pos_df['w_count']/pos_df['s_count']
#负样本
negfilename='G:\\sufe\\mental illness\\Dataset\\labeled\\negative\\data\\feature_count\\neg_count.csv'
neg_df = pd.read_csv(negfilename)
cols = ['a_count','c_count','me_count','you_count','him_count','CC','CD','IN','JJ','MD','NN','RB','RP','VB','WDT','WRB']
for item in cols:
    neg_df[item] /= neg_df['w_count']
neg_df['w_count']=neg_df['w_count']/neg_df['s_count']
#cols = ['a_count','c_count','me_count','you_count','him_count','w_count']
res_lev = [stats.levene(np.array(pos_df[i]), np.array(neg_df[i])) for i in cols]
res=[]
for i in range(len(cols)):
    if list(res_lev[i])[1]>0.05:
        res_ttest = stats.ttest_ind(np.array(pos_df[cols[i]]),np.array(neg_df[cols[i]]), equal_var = True)
    else:
        res_ttest = stats.ttest_ind(np.array(pos_df[cols[i]]),np.array(neg_df[cols[i]]), equal_var = False)
    res.append((cols[i],np.mean(pos_df[cols[i]]),np.mean(neg_df[cols[i]]),list(res_ttest)[1]/2))
w_lev = stats.levene(np.array(pos_df['w_count']), np.array(neg_df['w_count']))    
if list(w_lev)[1]>0.05:
    res_ttest = stats.ttest_ind(np.array(pos_df['w_count']),np.array(neg_df['w_count']), equal_var = True)
else:
    res_ttest = stats.ttest_ind(np.array(pos_df['w_count']),np.array(neg_df['w_count']), equal_var = False)
res.append(('w_count',np.mean(pos_df['w_count']),np.mean(neg_df['w_count']),list(res_ttest)[1]/2))                                               
#for item in res:
    #if item[3]<0.05:
        #print(item)
    #else:
        #pass
for item in res:
    print(item)
```

```python
import numpy as np
import pandas as pd
from scipy import stats
#正样本
posfilename='G:\\sufe\\mental illness\\Dataset\\labeled\\positive\\data\\infect_user_data.csv'
pos_df = pd.read_csv(posfilename)

pos_df["statuses_count"] = pos_df["statuses_count"] / pos_df["days"]
pos_df["followers_count"] = pos_df["followers_count"] / pos_df["days"]
pos_df["favourites_count"] = pos_df["favourites_count"] / pos_df["days"]

#负样本
negfilename='G:\\sufe\\mental illness\\Dataset\\labeled\\negative\\data\\negative_user_data.csv'
neg_df = pd.read_csv(negfilename)

neg_df["statuses_count"] = neg_df["statuses_count"] / neg_df["days"]
neg_df["followers_count"] = neg_df["followers_count"] / neg_df["days"]
neg_df["favourites_count"] = neg_df["favourites_count"] / neg_df["days"]
cols=['statuses_count','followers_count','favourites_count','friends_count']
res_lev = [stats.levene(np.array(pos_df[i]), np.array(neg_df[i])) for i in cols]
res=[]
for i in range(len(cols)):
    if list(res_lev[i])[1]>0.05:
        res_ttest = stats.ttest_ind(np.array(pos_df[cols[i]]),np.array(neg_df[cols[i]]), equal_var = True)
    else:
        res_ttest = stats.ttest_ind(np.array(pos_df[cols[i]]),np.array(neg_df[cols[i]]), equal_var = False)
    res.append((cols[i],np.mean(pos_df[cols[i]]),np.mean(neg_df[cols[i]]),list(res_ttest)[1]/2))
for item in res:
    print(item)
```
