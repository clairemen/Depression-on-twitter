```python
%matplotlib inline
import sys
import datetime
import matplotlib.pyplot as plt

import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
N=3
color_mine = ["#F8414A", "#5676A1"]

p_df = pd.read_csv("G:\\sufe\\mental illness\\Dataset\\labeled\\positive\\data\\pos_inter.csv")
n_df = pd.read_csv("G:\\sufe\\mental illness\\Dataset\\labeled\\negative\\data\\neg_inter.csv")

p_df["replay_count"] = p_df["replay_count"] / p_df["tweet_count"]
p_df["reweet_count"] = p_df["reweet_count"] / p_df["tweet_count"]

n_df["replay_count"] = n_df["replay_count"] / n_df["tweet_count"]
n_df["reweet_count"] = n_df["reweet_count"] / n_df["tweet_count"]


labels = ['positive','negative']
patches = [mpatches.Patch(color=color_mine[i], label="{:s}".format(labels[i]) ) for i in range(len(color_mine))]
fig,ax=plt.subplots(1,N,sharex=True,sharey=False,figsize=(15,4))
attributes_all = ["replay_count","reweet_count","tweet_count"]

x_value = [[0,1]] *N
#y_value = [[np.mean(p_df[attributes_all[i]]),np.mean(n_df[attributes_all[i]])] for i in range(N)]
y_value = [[np.median(p_df[attributes_all[i]]),np.median(n_df[attributes_all[i]])] for i in range(N)]
for x,y,sax,item in zip(x_value,y_value,ax,attributes_all):
    sax.bar(x,y,0.5,color=color_mine)
    sax.set_xticks(np.arange(2))
    sax.set_xticklabels(labels)
    sax.set_title('#'+ item)
ax[0].legend(handles=patches, bbox_to_anchor=(2,1.2), ncol=2)
#plt.tight_layout() 
plt.show()
```
